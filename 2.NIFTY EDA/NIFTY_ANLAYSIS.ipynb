## EDA of Nifty Data

```python
import pandas as pd
import numpy as np
import plotly.express as px
import chart_studio.plotly as py
import plotly.graph_objs as go
from plotly.offline import iplot
import cufflinks

cufflinks.go_offline()
cufflinks.set_config_file(world_readable=True, theme='pearl')

from datetime import datetime
```

```python
# Read the Nifty 50 data
nifty_50 = pd.read_csv('D:\\PROJECTS\\NIFTY\\INPUT\\NIFTY 50.csv')
nifty_50.head()
```

```python
nifty_50.info()
```

```python
# Check if any missing values.
nifty_50.isnull().sum()
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(
    x=nifty_50['Date'],
    y=nifty_50['High'],
    name='High Price',
line = dict(color='black'),
opacity = 0.8
))

fig.add_trace(go.Scatter(
    x=nifty_50['Date'],
    y=nifty_50['Low'],
    name='Low Price',
line = dict(color='red'),
opacity = 0.8
))

fig.update_layout(title_text='NIFTY 50 High vs Close Trend',plot_bgcolor='rgb(250, 242, 242)',yaxis_title='Value')

fig.show()
```

### Performance of other Nifty Sectoral Indicies in 2020

```python

nifty_auto = pd.read_csv('D:\\PROJECTS\\NIFTY\\INPUT\\NIFTY AUTO.csv',parse_dates=["Date"])
nifty_bank = pd.read_csv('D:\\PROJECTS\\NIFTY\\INPUT\\NIFTY BANK.csv',parse_dates=["Date"])
nifty_fmcg = pd.read_csv('D:\\PROJECTS\\NIFTY\\INPUT\\NIFTY FMCG.csv',parse_dates=["Date"])
nifty_IT = pd.read_csv('D:\\PROJECTS\\NIFTY\\INPUT\\NIFTY IT.csv',parse_dates=["Date"])
nifty_metal = pd.read_csv('D:\\PROJECTS\\NIFTY\\INPUT\\NIFTY METAL.csv',parse_dates=["Date"])
nifty_pharma = pd.read_csv('D:\\PROJECTS\\NIFTY\\INPUT\\NIFTY PHARMA.csv',parse_dates=["Date"])
```

```python
#fill in missing values
nifty_auto.fillna(method='ffill',inplace=True)
nifty_bank.fillna(method='ffill',inplace=True)
nifty_fmcg.fillna(method='ffill',inplace=True)
nifty_IT.fillna(method='ffill',inplace=True)
nifty_metal.fillna(method='ffill',inplace=True)
nifty_pharma.fillna(method='ffill',inplace=True)
```

```python
nifty_pharma.head()
```

### Compare closing prices of different secoral indices

```python
nifty_auto_2020 = nifty_auto[nifty_auto['Date'] > '2019-12-31']
nifty_bank_2020 = nifty_bank[nifty_bank['Date'] > '2019-12-31']
nifty_fmcg_2020 = nifty_fmcg[nifty_fmcg['Date'] > '2019-12-31']
nifty_IT_2020 = nifty_IT[nifty_IT['Date'] > '2019-12-31']
nifty_metal_2020 = nifty_metal[nifty_metal['Date'] > '2019-12-31']
nifty_pharma_2020 = nifty_pharma[nifty_pharma['Date'] > '2019-12-31']

```

```python
## Create a dictionary of closing values

d = {'Nifty Auto' : nifty_auto_2020['Close'].values,
     'Nifty Bank' : nifty_bank_2020['Close'].values,
     'Nifty FMCG' : nifty_fmcg_2020['Close'].values,
     'Nifty IT' : nifty_IT_2020['Close'].values,
     'Nifty Metal' : nifty_metal_2020['Close'].values,
     'Nifty Pharma' : nifty_pharma_2020['Close'].values
    }
```

```python
df = pd.DataFrame(data=d)
df.index = nifty_auto_2020['Date']
df.head()
```

```python
df.iplot(kind='box')
```

```python
fig = df.iplot(asFigure=True, subplots=True, subplot_titles=True, legend=False)
fig.show()
```

## Better Performing indices in 2020 , Nifty Pharma Index

```python
df_a=df['Nifty Pharma']
max_val=df_a.max()
min_val=df_a.min()
max_date=df_a[df_a==max_val].index[0].strftime('%Y-%m-%d')
min_date=df_a[df_a==min_val].index[0].strftime('%Y-%m-%d')
shape1=dict(kind='line',x0=max_date,y0=max_val,x1=min_date,y1=min_val,color='blue',width=2)
shape2=dict(kind='rect',x0=max_date,x1=min_date,fill=True,color='gray',opacity=.3)

df_a.iplot(shapes=[shape1,shape2])
```
